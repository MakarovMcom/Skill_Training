## 1. Просмотр датасета

```python
df = pd.read_csv('1.csv')
```

## 2. Чистка датасета

```python
# переименуем столбцы
df = df.rename(columns = {'vuokra': 'price',
                         'ala': 'description',
                         'osoite': 'address',
                         'ala-huoneet': 'long_d'})
```

```python
# удалим пробелы
df['price'] = df['price'].str.replace(" ", "")
df['description'] = df['description'].str.replace(" ", "")
df['address'] = df['address'].str.replace(" ", "")
df['long_d'] = df['long_d'].str.replace(" ", "")
```

```python
df.tail()
```

```python
df.head()
```

```python
df['price'] = df['price'].str.replace(",", "")
df['price'] = df['price'].str.replace("€/m", "")
```

#### столбец description

```python
# столбец description разделим на части
description_short = df["description"].str.split(",", n = 6, expand = True)
```

```python
description_short.sample(3)
```

```python
# разделим на чисти столбец 1 (квадратные метры и планировка)
description_short[['A', 'B']] = description_short[1].str.split('|', 2, expand=True)
```

```python
description_short.sample(3)
```

```python
description_short['A'] = description_short['A'].str.replace("m²", "")
description_short['A'] = description_short['A'].astype('float64') 
description_short

# переименуем столбцы
description_short = description_short.rename(columns = {'0': 'type',
                         'A': 'Sq',
                         'B': 'plan'})
```

```python
room_composition = description_short['plan'].str.split("+", n = 6, expand = True)
room_composition.sample(3)
```

```python
# соеденим описание и планировку
ds = description_short.assign(
    rooms = room_composition[0], 
    option1 = room_composition[1],
    option2 = room_composition[2],
    option3 = room_composition[3],
    option4 = room_composition[4],
    option5 = room_composition[5])
```

```python
# оставим нужные столбцы (тип дома, метры, планировка)
ds = ds.loc[:,[0,'Sq','rooms','option1','option2','option3','option4','option5']]
ds.head(3)
```

#### столбец address

```python
# удалим в адресе название города Эспоо
df['address'] = df['address'].str.replace("Espoo\n,", "")
df.head(3)
```

```python
# в адресе выделим улицу и район
df[['district', 'street']] = df['address'].str.split('\n\n,', 2, expand=True)

# оставим нужные столбцы (тип дома, метры, планировка)
df = df.loc[:,['price','district','street']]
df.head(3)
```

#### соеденим адрес с ценой и планировку с метражем

```python
# соеденим описание и планировку
df = df.assign(type_f = ds[0], 
               SQ = ds['Sq'], 
               rooms = ds['rooms'],
               opt1 = ds['option1'],
               opt2 = ds['option2'],
               opt3 = ds['option3'],
               opt4 = ds['option4'],
               opt5 = ds['option5'])
```

```python
df.info()
```

```python
df['price'] = df['price'].astype('float64') 
```

```python
# рассчитаем стоимость квадратного метра
df['price_meter'] = df['price'] / df['SQ']

df.head(3)
```

```python
# есть сауна 1, нет сауны 0
df['sauna'] = 'no'
df.loc[df['opt2'] == 's','sauna'] = 'sauna'
df.loc[df['opt4'] == 's','sauna'] = 'sauna'
df.head(3)
```

```python
# есть балкон 1 нет балкона 0
# 'prvk','parv' - 3
# 'parveke' - 2
# 'parv' - 4

df['balcony'] = 'no'
df.loc[df['opt2'] == 'parveke','balcony'] = 'parveke'
df.loc[df['opt3'] == 'prvk','balcony'] = 'parveke'
df.loc[df['opt3'] == 'parv','balcony'] = 'parveke'
df.loc[df['opt3'] == 'parv','balcony'] = 'parveke'
df.head(3)
```

```python
# есть остекление у балкона 2 или нет 0
# 'lasitettup' - 4
# 'lp' - 5

df.loc[df['opt4'] == 'lasitettup','balcony'] = 'lasitettup'
df.loc[df['opt4'] == 'lasitettuteras','balcony'] = 'lasitettup'
df.loc[df['opt5'] == 'lp','balcony'] = 'lasitettup'
df.head(3)
```

## 3. Анализ полученных данных

#### оценим районы

```python
# посмотрим на количество объявлений по районам и медианную стоимость квадратного метра
report = pd.pivot_table(df, 
                        index = 'district', 
                        values='price_meter', 
                        aggfunc={'count','median'}).reset_index()#.sort_values(by = 'median', ascending = False)
report['median'] = np.round(report['median'],2)
report
```

```python
# посчитаем разницу в процентах
max_price_meter = report['median'].max()
list_price = []
for i in report['median']:
    a = np.round(((i - max_price_meter)/max_price_meter) * 100,2)
    list_price.append(a)
print(list_price)
```

```python
# соеденим фреймы
report = report.assign(diff_pr = list_price)

# переименуем столбцы
report = report.rename(columns = {'column': 'num',
                         'median': 'price_meter',
                         'diff_pr': 'diff_in_pct'}).sort_values(by = 'price_meter', ascending = False)
report
```

```python
fig, ax = plt.subplots(figsize=(15,6))
ax.set_title("Стоимость аренды жилья в Еспоо [€/m²]");

# create dataset
height = report['price_meter']
bars = report['district']
y_pos = np.arange(len(bars))
 
# Create horizontal bars
plt.barh(y_pos, height)
 
# Create names on the x-axis
plt.yticks(y_pos, bars)

for i, v in enumerate(height):
    ax.text(v , i , str(v), color='black')

# Show graphic
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,8))
ax.set_title("Разброс цен на аренду квартир по районам [€/m²]");
sns.boxplot(y=df["district"], x=df["price_meter"])
plt.show()
```

```python
# цена аренды в топ районах
rich_district_price = df[(df['district'] == 'Kilo')|(df['district'] == 'Kuninkainen')]['price_meter'].median()
rich_district_price
```

```python
# цена аренды в остальных
other_district_price = df[(~(df['district'] == 'Kilo')|(df['district'] == 'Kuninkainen'))]['price_meter'].median()
other_district_price
```

```python
price_dif = (rich_district_price - other_district_price)/other_district_price
print('Цена квартиры в дорогих районах выше на {:.2%} по сравнению с остальными районами'.format(price_dif))
```
