## 2.1. Чистка датасета

```python
# переименуем столбцы
df = df.rename(columns = {'Column 1': 'price',
                         'Column 2': 'type',
                         'Column 3': 'arrangement',
                         'Column 4': 'address',
                         'Column 5':'availability'})
```

```python
# удалим пробелы
df['price'] = df['price'].str.replace(" ", "")
df['type'] = df['type'].str.replace(" ", "")
df['arrangement'] = df['arrangement'].str.replace(" ", "")
df['address'] = df['address'].str.replace(" ", "")
df['availability'] = df['availability'].str.replace(" ", "")
```

```python
df.tail()
```

```python
df.head()
```

```python
df['price'] = df['price'].str.replace("€/m", "")
df['price'] = df['price'].str.replace(",", "")
```

#### столбец arrangement

```python
# столбец arrangement разделим на части
df['arrangement'] = df['arrangement'].str.lower()
description_short = df["arrangement"].str.split("+", n = 6, expand = True)
```

#### столбец address

```python
# удалим в адресе название города Эспоо
df['address'] = df['address'].str.replace("Espoo\n,", "")
df.head(3)
```

```python
# в адресе выделим улицу и район
df[['district', 'street']] = df['address'].str.split('\n\n,', 2, expand=True)
df
```

#### столбец type

```python
# удалим в адресе название города Эспоо
df['type'] = df['type'].str.replace("m²", "")
```

```python
# разделим на чисти столбца type (квадратные метры и тип дома)
df[['flat_type', 'SQ']] = df['type'].str.split(',', 2, expand=True)
df
```

#### итоговый df

```python
# оставим нужные столбцы в датафрейме
df = df.loc[:,['price','district','street','flat_type','SQ']]
df['price'] = df['price'].astype('float64')  
df['SQ'] = df['SQ'].astype('float64') 
df['price_meter'] = df['price'] / df['SQ']
df = df.drop_duplicates()
df.head(3)
```

```python
df.info()
```

## 3.1. Анализ полученных данных

#### a) оценим районы с точки зрения цены за метр

```python
# посмотрим на количество объявлений по районам и медианную стоимость квадратного метра
report = pd.pivot_table(df, 
                        index = 'district', 
                        values='price_meter', 
                        aggfunc={'count','median'}).reset_index()#.sort_values(by = 'median', ascending = False)
report['median'] = np.round(report['median'],2)
report
```

```python
# посчитаем разницу в процентах
max_price_meter = report['median'].max()
list_price = []
for i in report['median']:
    a = np.round(((i - max_price_meter)/max_price_meter) * 100,2)
    list_price.append(a)
print(list_price)
```

```python
# соеденим фреймы
report = report.assign(diff_pr = list_price)

# переименуем столбцы
report = report.rename(columns = {'column': 'num',
                         'median': 'price_meter',
                         'diff_pr': 'diff_in_pct'}).sort_values(by = 'price_meter', ascending = False)
report
```

```python
fig, ax = plt.subplots(figsize=(15,25))
ax.set_title("Стоимость аренды жилья в Еспоо [€/m²]");

# create dataset
height = report['price_meter']
bars = report['district']
y_pos = np.arange(len(bars))
 
# Create horizontal bars
plt.barh(y_pos, height)
 
# Create names on the x-axis
plt.yticks(y_pos, bars)

for i, v in enumerate(height):
    ax.text(v , i , str(v), color='black')

# Show graphic
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,8))
ax.set_title("Разброс цен на аренду квартир по районам [€/m²]");
sns.boxplot(y=df["district"], x=df["price_meter"])
plt.show()
```

#### b) оценим районы с точки зрения цены за квартиру

```python
# посмотрим на количество объявлений по районам и медианную стоимость квадратного метра
report_flat = pd.pivot_table(df, 
                        index = 'district', 
                        values='price', 
                        aggfunc={'count','median'}).reset_index().sort_values(by = 'median', ascending = False)
report_flat
```

```python
fig, ax = plt.subplots(figsize=(15,8))
ax.set_title("Разброс цен на аренду квартир по районам [€/flat]");
sns.boxplot(y=df["district"], x=df["price"])
plt.show()
```

## 2.1. Просмотр датасета espoo_last

```python
df_n = pd.read_csv('espoo_last.csv', encoding = "ISO-8859-1")
df_n.head()
```

```python
df_n.info()
```

```python
north = 60.299
south = 60.129
east = 24.841
west = 24.585
```

```python
in_range = df_n[(df_n['latitude'] <= north)&(df_n['latitude'] >= south)&(df_n['longitude'] <= east)&(df_n['longitude'] >= west)]
out_range = df_n[~((df_n['latitude'] <= north)&(df_n['latitude'] >= south)&(df_n['longitude'] <= east)&(df_n['longitude'] >= west))]
```

```python
in_range.info()
```

```python
out_range.info()
```

```python
district_in_lat = pd.pivot_table(in_range, index = ['district'], values=['latitude',], aggfunc='median').reset_index()
district_in_lat.head(50)
```

```python
district_in_lon = pd.pivot_table(in_range, index = ['district'], values=['longitude',], aggfunc='median').reset_index()
district_in_lon.head(50)
```

```python
merge_one = out_range.merge(district_in_lat, left_on='district', right_on='district')
merge_two = merge_one.merge(district_in_lon, left_on='district', right_on='district')
merge_two.head()
```

```python
merge_two['latitude_x'] = merge_two['latitude_y']
merge_two['longitude_x'] = merge_two['longitude_y']
merge_two.head()
```

```python
# оставим нужные столбцы в датафрейме
merge_two = merge_two.loc[:,'description':'sauna']
merge_two
```

```python
# переименуем столбцы
merge_two = merge_two.rename(columns = {'latitude_x': 'latitude',
                         'longitude_x':'longitude'})
```

```python
export = pd.concat([in_range, merge_two])
```
